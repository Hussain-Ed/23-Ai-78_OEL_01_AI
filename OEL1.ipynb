ðŸ”¹ Part 3: Implement BFS

```python
import collections as cl
```

```python
def BFS(roads, start):
  shelterID = set()
  queue = cl.deque()
  while queue:
    start = queue.popleft()
    if start not in shelterID:
      shelterID.add(start)
      roads.extend(roads[start])
  return shelterID

print(BFS(roads, 'S1'))
```
